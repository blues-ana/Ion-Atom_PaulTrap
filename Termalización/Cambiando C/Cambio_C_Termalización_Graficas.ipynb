```python
import pandas as pd
import matplotlib.pyplot as plt

# Corregir la ruta en Windows
df = pd.read_csv(r'C:\Users\ANA NOGUERA\Desktop\Proyecto\Cambiando C\C_56.csv')

df.columns = ['energia_cinetica']

valor_maximo = 15
df_filtrado = df[df['energia_cinetica'] < valor_maximo]

x = range(1, len(df_filtrado) + 1)
y = df_filtrado['energia_cinetica']

valor_promedio = y.mean()

plt.figure(figsize=(10, 10))
plt.plot(x, y, color='blue')
plt.text(0.1, 0.95, '(b)', transform=plt.gca().transAxes, fontsize=24, verticalalignment='top', horizontalalignment='left')
plt.title('q = 0.1', fontsize=24)  # Title with reference label and larger font size
plt.ylabel('', fontsize=26)
plt.xlabel('Number of collisions', fontsize=26)
plt.xticks(fontsize=22)
plt.yticks(fontsize=22)
plt.axhline(valor_promedio, color='black', linestyle='--', label=f': {valor_promedio:.2f} K')
plt.legend(loc='lower right', fontsize=24)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el archivo CSV
df = pd.read_csv(r'C:\Users\ANA NOGUERA\Desktop\Proyecto\Cambiando Q\Q_0.1.csv')

# Asignar nombre a la columna
df.columns = ['energia_cinetica']

# Filtrar los datos para limitar el valor mximo
valor_maximo = 15
df_filtrado = df[df['energia_cinetica'] < valor_maximo]

# Graficar la distribucin usando KDE (estimacin de la densidad del kernel)
plt.figure(figsize=(10, 6))
plt.text(0.1, 0.95, '(b)', transform=plt.gca().transAxes, fontsize=24, verticalalignment='top', horizontalalignment='left')
sns.kdeplot(df_filtrado['energia_cinetica'], color='blue', linewidth=2)

# Ttulo y etiquetas
plt.title('q = 0.1', fontsize=24)  # Title with reference label and larger font size
plt.xlabel('Ion temperature (K)', fontsize=16)
plt.ylabel('Frequency', fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Mostrar la grfica
plt.show()
```
